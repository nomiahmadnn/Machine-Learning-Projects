```python
dataset.shape
```

```python
null_per = dataset.isnull().sum()/dataset.shape[0]*100
```

```python
col_to_drop = null_per[null_per>50].keys()
```

```python
col_to_drop
```

```python
dataset = dataset.drop(columns=col_to_drop)
```

```python
dataset.shape
```

```python
# Columns with null values
col_null_val = dataset.columns[dataset.isnull().any()]
col_null_val
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
numerical_col = dataset[col_null_val].select_dtypes(include=['int64', 'float64']).columns
numerical_col
```

### Add columns mean to numerical column

```python
for var in numerical_col:
    dataset[var] = dataset[var].fillna(dataset[var].mean())
```

```python
#dataset['LotFrontage'] = dataset['LotFrontage'].fillna(dataset['LotFrontage'].mean())
#dataset['MasVnrArea'] = dataset['MasVnrArea'].fillna(dataset['MasVnrArea'].mean())
#dataset['GarageYrBlt'] = dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].mean())
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
categorical_col = dataset[col_null_val].select_dtypes(include=['object']).columns
categorical_col
```

```python
# add columns mode for categorical column
len(categorical_col)
```

```python
for var in categorical_col:
    dataset[var] = dataset[var].fillna(dataset[var].mode()[0])
```

## Splittiing the dataset

```python
# Independent Variable
x = dataset.drop(columns='SalePrice')
```

```python
# Dependent Variable
y = dataset['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Building the Model

```python
# 1: Multiple Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regressor_mlr = LinearRegression()
regressor_mlr.fit(x_train, y_train)
```

```python
y_pred = regressor_mlr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
# 2: Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor_rf = RandomForestRegressor()
regressor_rf.fit(x_train, y_train)
```

```python
y_pred = regressor_rf.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
# 3: XGBoost Regressor
from xgboost import XGBRFRegressor
```

```python
regressor_xgb = XGBRFRegressor()
regressor_xgb.fit(x_train, y_train)
```

```python
y_pred = regressor_xgb.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
